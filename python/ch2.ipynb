```python
#結合之前所學用函數、迴圈寫出上圖程式(正比)
def func_1(x):
    return 2*x
x = []
y = []
for i in range (-10,10):
    x.append(i)
    y.append(func_1(i))
    
plt.plot(x,y)
plt.axis('equal')
plt.grid(color = '0.8')
plt.show()
```

```python
#numpy模組讓程式簡潔清爽
import numpy as np

def func_7a(x):
    return 2*x
def func_7b(x):
    return 2*x-9

x1 = np.arange(-1.5,2)
y1 = func_7a(x1)
plt.plot(x1,y1,color='chocolate',linewidth = '2', linestyle=':', marker='o')

x2 = np.arange(3,7)#產出給定區間內的數值
y2 = func_7b(x2)
plt.plot(x2,y2,color='deeppink',linewidth = '1', linestyle='--')

plt.grid(color='0.8')
plt.show()
    
```

```python
#解聯立方程式
import sympy as sp #匯入模組

x = sp.Symbol("x")
y = 2100-70*x
y
```

```python
ans = y.subs(x,10)#把10帶入y所指的運算式的x中,換句話說,以x = 10進行運算
ans
```

```python
#圓面積公式
#建立公式
r = sp.Symbol("r")
expr = sp.pi * r**2
display(expr)
#以pi = 3.14, r = 5帶入世子去做運算
ans = expr.subs({sp.pi:3.14, r:5})
ans
```
